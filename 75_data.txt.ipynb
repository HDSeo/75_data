```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = open("75_data.txt","w")
with open("raw_data.txt","r") as f:
    flag=""
    for line in f:
        
        a = line
        b = a.find("gamma")
        c = a.find("e-")
        
        if b!=-1:
                flag = "gamma"
            
        if c!=-1:
                flag = "e-"
        
        if flag=="gamma":
            try:
                if  a.split()[3]=="-75":
                    w.write(a)
            except:
                pass
                
            

w.close()      
```
